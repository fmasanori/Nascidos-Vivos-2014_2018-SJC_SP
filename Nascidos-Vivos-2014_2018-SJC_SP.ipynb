```python
# Desenvolvido por : Toni William do Prado
# O Pandas é uma biblioteca do Python utilizada para análise e manipulação de dados e nos permite trabalhar de forma rápida e
#     eficiente com arquivos do tipo csv,excel,txt,etc.
import pandas as pd
# O matplotlib é uma biblioteca do Python para criação de gráficos em 2D, bastante utilizada para visualização de dados e 
#     que apresenta uma série de possibilidades gráficas
import matplotlib.pyplot as plt
```

```python
# Abrindo um data frame para cada arquivo .csv
# Aqui temos 6 arquivos, onde os primeiros 5 são referentes aos anos de 2014 à 2018 e o último é a junção dos 5 últimos
#    disponíveis no DATASUS // TABNET

df14 = pd.read_csv('Nascimento por Mes 2014-2018\\nascimentoMes2014.csv', encoding = 'ANSI', delimiter=";", header=3)
df15 = pd.read_csv('Nascimento por Mes 2014-2018\\nascimentoMes2015.csv', encoding = 'ANSI', delimiter=";", header=3)
df16 = pd.read_csv('Nascimento por Mes 2014-2018\\nascimentoMes2016.csv', encoding = 'ANSI', delimiter=";", header=3)
df17 = pd.read_csv('Nascimento por Mes 2014-2018\\nascimentoMes2017.csv', encoding = 'ANSI', delimiter=";", header=3)
df18 = pd.read_csv('Nascimento por Mes 2014-2018\\nascimentoMes2018.csv', encoding = 'ANSI', delimiter=";", header=3)
dfT = pd.read_csv('Nascimento por Mes 2014-2018\\nascimentoMes2014-2018.csv', encoding = 'ANSI', delimiter=";", header=3)
```

```python
# Coletando informações básicas sobre o DataFrame:

# O DataFrame tem muitas linhas de dados, pra visualizar sucintamente as primeiras linhas de um DataFrame
#    existe o método .head()

df14.head()
```

```python
# Quantidade de linhas e colunas do DataFrame
df14.shape
```

```python
# Foi Observado que nas células da coluna do Município é composto pelo código IBGE mais o nome do município
# Pesquisando a linha no DataFrame através da coluna Município
df14[df14.Município ==  "354990 São José dos Campos"]
```

```python
# Selecionando os valores através do index descoberto com a pesquisa anterior
# Retorna um objeto contendo as colunas e o valor correspondente da linha do index selecionado
# Exemplo: { 'Município':'354990 São José dos Campos', 'Janeiro':'900' ... }
df14.loc[559]
```

```python
# Aqui começamos a coletar os valores de janeiro à dezembro do respectivo ano e guardamos em um array

# 2014
qntNascimentos14 = []
for i in range(1,13):
    qntNascimentos14.append(int(df14.loc[559][i]))
    
# 2015
qntNascimentos15 = []
for i in range(1,13):
    qntNascimentos15.append(int(df15.loc[559][i]))
    
# 2016
qntNascimentos16 = []
for i in range(1,13):
    qntNascimentos16.append(int(df16.loc[559][i]))
    
# 2017
qntNascimentos17 = []
for i in range(1,13):
    qntNascimentos17.append(int(df17.loc[559][i]))
    
# 2018
qntNascimentos18 = []
for i in range(1,13):
    qntNascimentos18.append(int(df18.loc[559][i]))
    
# 2014 - 2018
qntNascimentosT = []
for i in range(1,13):
    qntNascimentosT.append(int(dfT.loc[559][i]))
```

```python
# Foi definido um array para guardar os meses
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril',
        'Maio', 'Junho', 'Julho', 'Agosto',
        'Setembro', 'Outubro', 'Novembro', 'Dezembro']
```

```python
# Dados 2014
df14[df14.Município ==  "354990 São José dos Campos"]
```

```python
# Definindo o tamanho do gráfico
plt.figure(figsize = (20, 12))
# Formatando as labels do eixo x (Girando e Tamanho da fonte)
plt.xticks(rotation=35, fontsize=25)
# Formatando as labels do eixo y (Tamanho da fonte)
plt.yticks(fontsize=20)
# Formatando o título do gráfico (Definindo nome e Tamanho da fonte)
plt.title('Ano: 2014', fontsize=40)
# Declarando e formatando estilo do gráfico
# Definindo o array dos meses como eixo x e a quantidade de nascimentos no eixo y
# Cor da linha: preta, Cor do ponto: Laranja, Estilo do ponto: Diamante, Tamanho do Ponto: 15
plt.plot(meses, qntNascimentos14, color="k", mfc="orange", marker = 'D', ms=15)
```

```python
plt.figure(figsize = (20, 12))
plt.xticks(rotation=35, fontsize=25)
plt.yticks(fontsize=20)
plt.title('Ano: 2015', fontsize=40)
plt.plot(meses, qntNascimentos15, color="k", mfc="b", marker = 'D', ms=15)
```
